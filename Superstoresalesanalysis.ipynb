```python
df["Order_Date"]=pd.to_datetime(df["Order_Date"])
df["Ship_Date"]=pd.to_datetime(df["Ship_Date"])
```

```python
df["Postal_Code"]=df["Postal_Code"].astype(str)
```

```python
df["Month"]=df["Order_Date"].dt.month
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(" ","_")
```

Monthly_sales


```python
monthly_sales=df.groupby("Month")["Sales"].sum()
```

```python
plt.figure(figsize=(10,4))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o',color="Green")
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

Sales By Region

```python
sales_by_region=df.groupby("Region")["Sales"].sum()
```

```python
plt.figure(figsize=(4,4))
graph=sns.barplot(x=sales_by_region.index,y=sales_by_region.values)
plt.xlabel("Region")
plt.ylabel("Sales")
plt.title("Sales_by_Region")
# graph.labels(graph.containers[0])
plt.show()
```

Sales by Category

```python
sales_by_Category=df.groupby("Category")["Sales"].sum()
plt.figure(figsize=(4,4))
graph=sns.barplot(x=sales_by_Category.index,y=sales_by_Category.values)
plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Sales_by_Category")
# graph.labels(graph.containers[0])
plt.show()
```

Sales by Sub_Category


```python
sales_by_sub_Category=df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
plt.figure(figsize=(14,4))
graph=sns.barplot(x=sales_by_sub_Category.index,y=sales_by_sub_Category.values,color="red")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.title("Sales_by_Sub_Category")
# graph.labels(graph.containers[0])
plt.show()
```

Profit by Region

```python
profit_region = df.groupby('Region')['Profit'].sum()

plt.figure(figsize=(5,4))
sns.barplot(x=profit_region.index, y=profit_region.values)
plt.title('Profit by Region')

plt.show()
```

Profit vs Discount

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Discount",y="Profit",data=df)
plt.title("Discount vs Profit")
plt.show()
```

Top 10 Products by Sales

```python
top_products = df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top 10 Products by Sales')
plt.show()
```

Bottom 10 Products by Sales

```python
bottom_products = df.groupby('Product_Name')['Sales'].sum().sort_values().head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=bottom_products.values, y=bottom_products.index)
plt.title('Top 10 Products by Sales')
plt.show()
```

The business shows strong revenue growth driven by specific regions and products, but profitability is impacted by aggressive discounting and low-performing products.
